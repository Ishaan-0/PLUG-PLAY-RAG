```python
import os
import platform

def get_root_directory():
    system = platform.system()
    print(system)
    if system == "Windows":
        # On Windows, get the drive of the current directory
        current_drive = os.path.splitdrive(os.getcwd())[0]
        return current_drive + os.sep  # e.g., 'C:\'
    else:
        # On Unix-like systems (Linux, macOS), root is always '/'
        return '/'

# Usage
root = get_root_directory()
print(f"Root directory: {root}")
```

```python
import shutil
import os

directory_to_remove = "/Volumes/T7/Year_3_Docs/Sem_6/chroma_db_path"

if os.path.exists(directory_to_remove) and os.path.isdir(directory_to_remove):
    try:
        shutil.rmtree(directory_to_remove)
        print(f"Directory '{directory_to_remove}' removed successfully.")
    except OSError as e:
        print(f"Error removing directory '{directory_to_remove}': {e}")
else:
    print(f"Directory '{directory_to_remove}' does not exist or is not a directory.")
```
